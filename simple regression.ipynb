```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
df=pd.read_csv("Advertising Budget and Sales.csv")
df["total"]=df["TV Ad Budget ($)"]+df["Radio Ad Budget ($)"]+df["Newspaper Ad Budget ($)"]
df.head()
data=df[["Sales ($)","total"]].copy()
data.head()
```

```python
sn.scatterplot(data=data,x="total",y="Sales ($)")
```

```python
sn.regplot(data=data,x="total",y="Sales ($)")
```

```python
X=data["total"]
Y=data["Sales ($)"]
```

```python
#y=mx+b
help(np.polyfit)
```

```python
np.polyfit(X,Y,deg=1)
```

```python
spent=np.linspace(0,500,100)
prediction=0.04868788*spent+4.24302822

```

```python
sn.scatterplot(data=data,x="total",y="Sales ($)")
plt.plot(spent,prediction,color="green")
```

```python
#creating polynomial function
np.polyfit(X,Y,3)


```
